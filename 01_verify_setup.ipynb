# TensorFlow Setup Verification

Run each cell below to verify your installation is working correctly.

```python
# Cell 1: Import TensorFlow and check version
import tensorflow as tf
print(f"TensorFlow version: {tf.__version__}")
print(f"Keras version: {tf.keras.__version__}")
```

```python
# Cell 2: Test basic TensorFlow operations
# Create some tensors
a = tf.constant([[1, 2], [3, 4]])
b = tf.constant([[5, 6], [7, 8]])

# Matrix multiplication
result = tf.matmul(a, b)
print("Matrix multiplication result:")
print(result.numpy())
print("\nTensorFlow is working correctly!")
```

```python
# Cell 3: Test Keras (neural network library)
from tensorflow import keras
from keras import layers

# Create a simple model
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(10,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

model.summary()
print("\nKeras is working correctly!")
```

```python
# Cell 4: Test matplotlib for visualizations
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('Test Plot - Matplotlib Working!')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()

print("Matplotlib is working correctly!")
```
